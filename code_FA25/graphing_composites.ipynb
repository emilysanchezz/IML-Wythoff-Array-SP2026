```python
import math
import matplotlib.pyplot as plt
import numpy as np


global phi
phi = (1 + math.sqrt(5))/2

def a_n_plug(n):
    a_n = math.floor(n*phi)
    return a_n

def b_n_plug(n):
    b_n = math.floor(n*phi**2)
    return b_n

def composite(n, outter, inner):
    a_n = math.floor(n*phi)
    b_n = math.floor(n*phi**2)

    funct = None

    if (outter == 'a'):
        if (inner == 'b'):
            funct = a_n_plug(b_n)
        else:
            funct = a_n_plug(a_n)
    else:
        if (inner == 'a'):
            funct = b_n_plug(a_n)
        else:
            funct = b_n_plug(b_n)     

    return funct

def iterate(n, outter, inner, k):
    x=n
    for _ in range(k):
        x = composite(x, outter, inner)
    return x

N = 50
n_values = range(1, N+1)
max_iter = 5   # how many times to iterate each composite

#--------
composite_a_a = [
    ('a','a')
]

plt.figure(figsize=(12,8))

for (outter, inner) in composite_a_a:
    for k in range(1, max_iter+1):
        vals = [iterate(n, outter, inner, k) for n in n_values]
        label = f"{outter}({inner}(n)) iter^{k}"
        plt.plot(n_values, vals, markersize=2, label=label)

plt.xlabel("n")
plt.ylabel("Value after iterations")
plt.title("Iterating each composite multiple times")
plt.legend()
plt.grid(True)
plt.show()
#------
composite_a_b = [
    ('a','b')
]

plt.figure(figsize=(12,8))

for (outter, inner) in composite_a_b:
    for k in range(1, max_iter+1):
        vals = [iterate(n, outter, inner, k) for n in n_values]
        label = f"{outter}({inner}(n)) iter^{k}"
        plt.plot(n_values, vals, markersize=2, label=label)

plt.xlabel("n")
plt.ylabel("Value after iterations")
plt.title("Iterating each composite multiple times")
plt.legend()
plt.grid(True)
plt.show()
#----
composite_b_b = [
    ('b','b')
]

plt.figure(figsize=(12,8))

for (outter, inner) in composite_b_b:
    for k in range(1, max_iter+1):
        vals = [iterate(n, outter, inner, k) for n in n_values]
        label = f"{outter}({inner}(n)) iter^{k}"
        plt.plot(n_values, vals, markersize=2, label=label)

plt.xlabel("n")
plt.ylabel("Value after iterations")
plt.title("Iterating each composite multiple times")
plt.legend()
plt.grid(True)
plt.show()
#----
composite_b_a = [
    ('b','a')
]

plt.figure(figsize=(12,8))

for (outter, inner) in composite_b_a:
    for k in range(1, max_iter+1):
        vals = [iterate(n, outter, inner, k) for n in n_values]
        label = f"{outter}({inner}(n)) iter^{k}"
        plt.plot(n_values, vals, markersize=2, label=label)

plt.xlabel("n")
plt.ylabel("Value after iterations")
plt.title("Iterating each composite multiple times")
plt.legend()
plt.grid(True)
plt.show()
#----


n_values = np.linspace(0, 5, 100)
results = {
    'a(a(n))': [composite(n, 'a', 'a') for n in n_values],
    'a(b(n))': [composite(n, 'a', 'b') for n in n_values],
    'b(a(n))': [composite(n, 'b', 'a') for n in n_values],
    'b(b(n))': [composite(n, 'b', 'b') for n in n_values]
}

plt.figure(figsize=(8,6))

for label, y_vals in results.items():
    plt.plot(n_values, y_vals, label=label)

plt.xlabel('n')
plt.ylabel('Composite value')
plt.title('Behavior of Composite Functions')
plt.legend()
plt.grid(True)
plt.show()
#----

N = 20
n_values = range(1, N+1)

a_vals = [a_n_plug(n) for n in n_values]
b_vals = [b_n_plug(n) for n in n_values]
a_a_vals = [composite(n, 'a', 'a') for n in n_values]
b_a_vals = [composite(n, 'b', 'a') for n in n_values]
a_b_vals = [composite(n, 'a', 'b') for n in n_values]
b_b_vals = [composite(n, 'b', 'b') for n in n_values]

plt.figure(figsize=(10,6))
plt.plot(n_values, a_vals, 'o-', label='a(n)')
plt.plot(n_values, b_vals, 'o-', label='b(n)')
plt.plot(n_values, a_a_vals, 'x--', label='a(a(n))')
plt.plot(n_values, b_a_vals, 'x--', label='b(a(n))')
plt.plot(n_values, a_b_vals, 'x--', label='a(b(n))')
plt.plot(n_values, b_b_vals, 'x--', label='b(b(n))')

plt.xlabel('n')
plt.ylabel('Value')
plt.title('Wythoff Sequences and Composites')
plt.grid(True)
plt.legend()
plt.show()
```
